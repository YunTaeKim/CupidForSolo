```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBClassifier, plot_importance
from sklearn.model_selection import cross_val_score
from collections import Counter
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
import pickle

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

% matplotlib inline
```

## 남성 데이터

```python
man = pd.read_csv('man.csv', encoding = 'cp949')
man = man.drop_duplicates()
man = man.sample(frac = 1, random_state = 1024)
man.index = range(len(man))
```

```python
man_X = man.drop('target_solo', axis = 1).astype('float')
col_names = list(man_X.columns)
man_y = (man['target_solo'] == 'solo').astype('float')
```

```python
model = XGBClassifier()
model.fit(X = man_X, y = man_y)
val = cross_val_score(model, X = man_X, y = man_y, scoring = 'accuracy')
acc = np.mean(val)
print(acc)

fig, ax = plt.subplots(1, 1, figsize = (12, 8))
plot_importance(model, ax = ax)
plt.show()
```

```python
importance = pd.DataFrame([(n, v) for (n, v) in zip(col_names, model.feature_importances_)], columns = ['feature', 'value'])
importance = importance.sort_values('value', ascending = False)
```

```python
np.random.seed(10232)
idx = np.random.choice(range(len(man)), 20)
test_man = man.loc[idx]
man = man.drop(idx)
man_X = man.drop('target_solo', axis = 1).astype('float')
man_y = (man['target_solo'] == 'solo').astype('float')
test_man_X = test_man.drop('target_solo', axis = 1).astype('float')
test_man_y = (test_man['target_solo'] == 'solo').astype('float')
```

```python
Counter(man_y)
```

```python
Counter(test_man_y)
```

```python
scaler = MinMaxScaler()
clf = XGBClassifier()
model = Pipeline([('scaler', scaler), ('clf', clf)])

train = []
test = []
for i in range(1, 40, 1):
    col_n = importance.iloc[:i]['feature']
    temp_X = man_X[col_n]
    te_X = test_man_X[col_n]
    model.fit(X = temp_X, y = man_y)
    val = cross_val_score(model, X = temp_X, y = man_y, scoring = 'f1', cv = 5)
    pred = model.predict(te_X)
    real = test_man_y
    train_acc = np.mean(val)
    test_acc = f1_score(real, pred)
    train.append(train_acc)
    test.append(test_acc)
```

```python
plt.plot(range(1, 40, 1), train, label = 'train', marker = 'o')
plt.legend()
plt.show()
```

```python
col_n = list(importance.iloc[:8]['feature']) + ['glasses_안경착용', 'accessory_earring']
man_X = man_X[col_n]
test_man_X = test_man_X[col_n]
```

```python
col_n
```

```python
man_dict = {'train_X': man_X,
            'train_y': man_y,
            'test_X': test_man_X,
            'test_y': test_man_y}

file = open('man_data.pkl', 'wb')
pickle.dump(man_dict, file)
file.close()
```
